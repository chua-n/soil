```python
import os
from skimage import io
import numpy as np

from particle.utils.plotter import makeGrid
import matplotlib.pyplot as plt

import matplotlib as mpl
zhfont = mpl.font_manager.FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc')
# labelsize = 16
# plt.style.use("bmh")
```

```python
basefolder = "/home/chuan/inter/"
# basefolder = "/home/chuan/surround/"
folders = os.listdir(basefolder)
folders.sort(key=float)
# folders = folders[:3]
print(folders)
imgs = []
for folder in folders:
    filenames = os.listdir(os.path.join(basefolder, folder))
    filenames.sort()
    print(filenames)
    for filename in filenames:
        imgs.append(io.imread(os.path.join(basefolder, folder, filename)))
```

```python
saveName = "/home/chuan/imgGrid.png"
makeGrid(imgs, saveName, nrow=6)
```

```python
# for interpolate
img = io.imread(saveName)
plt.imshow(img)
xUnit = img.shape[1] // 6
yUnit = img.shape[0] // 5
plt.xticks(range(xUnit//2, img.shape[1], xUnit), [round(i*0.2, 1) for i in range(6)], size=8)
plt.yticks(range(yUnit//2, img.shape[0], yUnit), [f"试验{i}" for i in range(1, 6)], fontproperties=zhfont, size=8)
plt.xlabel(r"$\beta$")
# plt.subplots_adjust(left=0, bottom=0, right=1, top=1, hspace=0.1,wspace=0.1)
plt.tight_layout()
plt.savefig("/home/chuan/plot.png", dpi=300)
```
