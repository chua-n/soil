```python
import os
from skimage import io
import numpy as np

from particle.utils.plotter import makeGrid
import matplotlib.pyplot as plt

import matplotlib as mpl
zhfont = mpl.font_manager.FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc')
# labelsize = 16
# plt.style.use("bmh")
```

```python
# basefolder = "/home/chuan/interpolate/"
basefolder = "/home/chuan/surround/"
folders = os.listdir(basefolder)
folders.sort(key=float)
folders = folders[:3]
print(folders)
imgs = []
for folder in folders:
    filenames = os.listdir(os.path.join(basefolder, folder))
    filenames.sort()
    print(filenames)
    for filename in filenames:
        imgs.append(io.imread(os.path.join(basefolder, folder, filename)))
```

```python
saveName = "/home/chuan/imgGrid.png"
makeGrid(imgs, saveName, nrow=6)
```

```python
# for surround
img = io.imread(saveName)
plt.imshow(img)
xUnit = img.shape[1] // 6
yUnit = img.shape[0] // 3
plt.xticks(range(xUnit//2, img.shape[1], xUnit), 
           ["原颗粒"]+[rf"$\sigma={round(i*0.1, 1)}$" for i in range(1, 6)],
            fontproperties=zhfont, size=8)
plt.yticks(range(yUnit//2, img.shape[0], yUnit), [f"视角{i}" for i in range(1, 4)], 
           fontproperties=zhfont, size=8)
plt.savefig("/home/chuan/plot.png", dpi=300)
```
