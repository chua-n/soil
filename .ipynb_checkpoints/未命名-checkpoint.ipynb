```python
import numpy as np
from skimage import io, color, filters, measure, img_as_ubyte
from scipy import ndimage as ndi
import matplotlib.pyplot as plt
import napari

from particle.pipeline import SandHeap
```

```python
sandHeap = SandHeap(source=None,
                   se=ndi.generate_binary_structure(rank=3, connectivity=2), 
                   connectivity=1,
                   persistencePath='/home/chuan/',
                   cubeSize=64)
```

```python
%%time
sandHeap.preSegmented = np.load('./data/liutao/v1/pre-segmented.npy')
sandHeap.setStatus('pre-segmented')
```

```python
labeled, num = measure.label(sandHeap.preSegmented, return_num=True, connectivity=1)
print(num)
```

```python
%%time
regionprops = measure.regionprops(labeled)
```

```python
areas = [region.area for region in regionprops]
plt.violinplot(areas, showextrema=False)
```

```python
plt.hist(areas)
```

```python
plt.style.use('seaborn')
_, ax = plt.subplots(3, 1, sharex=True, figsize=(16, 12))
ax[0].plot(sorted(areas))
ax[1].plot(sorted(areas)[:10000])
ax[2].plot(range(len(areas)-10000, len(areas)), sorted(areas)[-10000:])
```
