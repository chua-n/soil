```python
import numpy as np
from particle.pipeline import SandHeap
from skimage import io, measure, img_as_ubyte
from scipy import ndimage as ndi
import napari
```

```python
sandHeap = SandHeap(source='./data/liutao/binary.npy',
                   persistencePath='./data/liutao/',
                   cubeSize=64)
```

```python
sandHeap.preSegmented = np.load('./data/liutao/pre-segmented-1-7.npy')
sandHeap.setStatus("pre-segmented")
```

```python
sandHeap.removeBigSegmentationFace('3d', 500, connectivity=2, persistence=False, returnDiagram=True)
```

```python
%%time
finalLabeled, num = measure.label(sandHeap.finalSegmented, return_num=True, connectivity=1)
print(num)
```

```python
%%time
preLabeled, num = measure.label(sandHeap.preSegmented, return_num=True, connectivity=1)
print(num)
```

```python
%gui qt
# napari.view_labels(finalLabeled[200:300, :600, :600])
napari.view_labels(preLabeled)
```
